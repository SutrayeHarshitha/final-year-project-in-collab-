```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the csv data to a Pandas Dataframe
heart_data = pd.read_csv('heart_cleveland.csv')
```

```python
# statistical measures about the data
heart_data.describe()
```

```python
# Checking the distribution of target variable (target values)
# (1) ---------> Defective heart
# (0) ---------> Healthy heart
heart_data['target'].value_counts()
```

```python
# Splitting the features and target values
x = heart_data.drop(columns='target',axis=1)
y = heart_data['target']
print(x)
```

```python
print(y)
```

```python
# splitting the data into training data and testing data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify = y,random_state = 2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
# logistic regression
model = LogisticRegression()
```

```python
# training the LogisticRegression model with the training data
model.fit(x_train,y_train)
```

```python
# accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print("Accuracy on Training data : ",training_data_accuracy)
```

```python
# accuracy on testing data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
print("Accuracy on Testing data : ",test_data_accuracy)
```
